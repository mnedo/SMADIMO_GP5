```python
def get_main_product_image_url(soup, base_url="https://oskelly.ru"):
    candidates = []

    for img in soup.find_all("img"):
        src = img.get("src")
        srcset = img.get("srcset")
        class_name = " ".join(img.get("class", []))

        is_product_gallery = (
            "osk-product-gallery__main-swiper-image" in class_name
            or "osk-product-gallery__thumbs-swiper-image" in class_name
            or "osk-product-card__card-img-pic" in class_name
        )

        if src:
            full_url = urljoin(base_url, src)

            if "/product/" in full_url and full_url.startswith("http"):
                if is_product_gallery:
                    return full_url
                candidates.append(full_url)

        if srcset:
            for part in srcset.split(","):
                srcset_url = part.strip().split(" ")[0]
                full_url = urljoin(base_url, srcset_url)

                if "/product/" in full_url and full_url.startswith("http"):
                    if is_product_gallery:
                        return full_url
                    candidates.append(full_url)

    if candidates:
        return candidates[0]

    return None
```

```python
def download_main_product_image(image_url, item_id, images_dir=images_dir):

    if image_url is None:
      return None

    ext = ".jpg"

    if ".png" in image_url.lower():
      ext = ".png"
    elif ".jpeg" in image_url.lower():
      ext = ".jpeg"

    image_path = os.path.join(images_dir, f"{item_id}{ext}")

    if os.path.exists(image_path):
      return image_path

    img_response = requests.get(image_url,headers={"User-Agent": headers["User-Agent"],"Referer": "https://oskelly.ru"},timeout=20)

    if img_response.status_code == 200 and img_response.content:
      with open(image_path, "wb") as f:
          f.write(img_response.content)

      return image_path

    return None
```

```python
all_items = []

for i in range(0, 100):
  url = 'https://oskelly.ru' + items.iloc[i, 0]
  oskelly = requests.get(url, headers=headers)

  if oskelly.status_code == 200:
    soup = BeautifulSoup(oskelly.text, 'html')
    item = {}

    item_id = items.iloc[i, 0].split('-')[-1]
    item['Oskelly ID'] = item_id
    item['url'] = url

    for elem in soup.find_all('li', class_='osk-product-attributes__item'):
      name_elem = elem.find('span', class_='osk-product-attributes__item-name')
      value_elem = elem.find('span', class_='osk-product-attributes__item-value')

      if name_elem and value_elem:
        category = name_elem.text.strip()
        value = value_elem.text.strip()
        item[category] = value

    price = soup.find('span', class_='osk-product__rrpPrice')
    if price:
      item['Цена'] = price.text.replace('\xa0', '').replace('₽', '').strip()

    discount = soup.find('span', class_='osk-product__discount')
    if discount:
      item['Скидка'] = discount.text.strip()

    price2 = soup.find('span', class_='osk-product__price')
    if price2 and 'Цена' not in item:
      item['Цена'] = price2.text.replace('\xa0', '').replace('₽', '').strip()


    description = soup.find('div', class_='osk-product-seller__description')
    if description:
      item['Описание'] = description.text.strip()


    brand = soup.find('a', class_='osk-product__brand')
    if brand:
      item['Бренд'] = brand.text.strip()


    title = soup.find('h1', class_='osk-product__title')
    if title:
      item['Название'] = title.text.strip()


    condition = soup.find('div', class_='osk-product__condition')
    if condition:
      item['Состояние'] = condition.text.strip()


    image_url = get_main_product_image_url(soup)
    image_path = download_main_product_image(image_url=image_url,item_id=item_id,images_dir=images_dir)

    item['image_url'] = image_url
    item['image_path'] = image_path

    all_items.append(item)

    print(item_id, item.get('Бренд'), item.get('Цена'), image_path)

  time.sleep(0.5)
```

```python
data = pd.DataFrame(all_items)
data.to_excel('all_items_100_with_images.xlsx', index=False)
data.head()
```
